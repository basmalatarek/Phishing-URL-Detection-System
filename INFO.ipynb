```python
# -----------------------------
# Outlier Flagging using IQR
# -----------------------------
# The IQR method detects extreme values outside the normal range.
# Values lower than (Q1 - 1.5 * IQR) or higher than (Q3 + 1.5 * IQR) are considered outliers.

# Select only numeric columns
numeric_df = df.select_dtypes(include=['int64','float64'])

# Calculate quartiles
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Detect outliers
outliers = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR)))

# Count outliers per column
outlier_counts = outliers.sum()

# Sort columns by number of outliers
outlier_counts.sort_values(ascending=False)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

# 1.3.3.2. Multicollinearity Check using VIF
# We pick key features that showed high correlation earlier
check_cols = ['page_rank', 'domain_age', 'length_url', 'nb_dots', 'google_index', 'nb_hyperlinks']

# Drop any remaining NaNs only for this calculation to avoid errors
vif_df = df[check_cols].dropna()

vif_data = pd.DataFrame()
vif_data["feature"] = check_cols

# Calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(vif_df.values, i) for i in range(len(check_cols))]

print("--- Multicollinearity Check (VIF) ---")
print(vif_data.sort_values(by="VIF", ascending=False))
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np

#  Principal Component Analysis (PCA)
# PCA is sensitive to scale, so we must standardize the data first
# We exclude the 'status' and 'url' columns
numeric_data = df.select_dtypes(include=['int64', 'float64']).drop(columns=['status'], errors='ignore')

# Handle any remaining NaNs after the cleaning steps to ensure PCA works
numeric_data_filled = numeric_data.fillna(numeric_data.median())

# Standardize the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data_filled)

# Apply PCA - Let's see how much variance is explained by all components
pca = PCA()
pca.fit(scaled_data)

# Calculate cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# Plot the Scree Plot (Cumulative Variance)
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.axhline(y=0.95, color='r', linestyle='-', label='95% Explained Variance')

plt.title('PCA: Cumulative Explained Variance', fontsize=14)
plt.xlabel('Number of Components', fontsize=12)
plt.ylabel('Cumulative Explained Variance', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()

# Print how many components are needed for 95% variance
n_95 = np.argmax(cumulative_variance >= 0.95) + 1
print(f"Number of components needed to explain 95% of variance: {n_95}")
```

## Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder

# Encoding Categorical Variables

# Re-confirming the encoding of 'status'
# Phishing = 1, Legitimate = 0
if df['status'].dtype == 'O' or df['status'].dtype == 'object':
    le = LabelEncoder()
    df['status'] = le.fit_transform(df['status'])
    print("Status column encoded successfully.")
else:
    print("Status column is already numeric.")

# Drop the 'url' column as it is not needed for the machine learning model
# We keep a copy of the cleaned dataframe before dropping
df_final = df.drop(columns=['url'], errors='ignore')

# Check the first few rows to ensure everything is numeric
print("\n--- Data Types after Encoding ---")
print(df_final[['status']].head())
print(df_final.dtypes.value_counts())
```

```python
from sklearn.preprocessing import RobustScaler
import pandas as pd

#  Feature Scaling

# Step A: Separate Features (X) and Target (y)
X = df_final.drop(columns=['status'])
y = df_final['status']

# Step B: Initialize the RobustScaler
# We use RobustScaler because it is less sensitive to outliers than StandardScaler
scaler = RobustScaler()

# Step C: Fit and transform the features
X_scaled = scaler.fit_transform(X)

# Step D: Convert back to a DataFrame to keep the column names
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

# Check the results
print("--- Features after Scaling (First 5 rows) ---")
print(X_scaled_df.head())
print("\nDescriptive statistics for a scaled feature (length_url):")
print(X_scaled_df['length_url'].describe())
```

```python
import numpy as np

#  Outlier Treatment using Capping (Winsorization)
# We will cap the extreme values at the 1st and 99th percentiles
# This keeps the "signal" of the outlier but limits its "damage" to the model

X_capped = X_scaled_df.copy()

for col in X_capped.columns:
    lower_limit = X_capped[col].quantile(0.01)
    upper_limit = X_capped[col].quantile(0.99)
    
    # Apply capping
    X_capped[col] = np.where(X_capped[col] < lower_limit, lower_limit,
                             np.where(X_capped[col] > upper_limit, upper_limit, X_capped[col]))

print("--- Outliers Capped Successfully ---")
# Check if the Max value of length_url decreased
print(f"New Max for length_url: {X_capped['length_url'].max():.2f}")
```

```python
#  Handle Imbalanced Data
# Since we checked in EDA that status is 50% Phishing and 50% Legitimate, 
# no further action (like SMOTE) is required.

print("--- Target Balance Check ---")
print(y.value_counts(normalize=True) * 100)
```

```python
from sklearn.model_selection import train_test_split

# Train & Test Split
# We will split the data into 80% Training and 20% Testing
X_train, X_test, y_train, y_test = train_test_split(X_capped, y, test_size=0.2, random_state=42, stratify=y)

print("--- Splitting Completed ---")
print(f"Training set size: {X_train.shape}")
print(f"Testing set size: {X_test.shape}")
```

## Models

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report, confusion_matrix

# Train classification models
# We will create a dictionary to hold our models for easy comparison
models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42)
}

#  Evaluation Metrics
for name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1] # Needed for ROC-AUC
    
    print(f"\n" + "="*30)
    print(f"Model: {name}")
    print("="*30)
    
    #  Accuracy
    acc = accuracy_score(y_test, y_pred)
    
    #  F1-Score
    f1 = f1_score(y_test, y_pred)
    
    #  ROC-AUC
    roc = roc_auc_score(y_test, y_prob)
    
    print(f"Accuracy: {acc*100:.2f}%")
    print(f"F1-Score: {f1:.4f}")
    print(f"ROC-AUC:  {roc:.4f}")
    
    #  Classification Report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

#  Optional: Confusion Matrix visualization for the best model (RF)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test, models["Random Forest"].predict(X_test)), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Random Forest')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

#  Plotting ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, models["Random Forest"].predict_proba(X_test)[:, 1])
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (Legitimate classified as Phishing)')
plt.ylabel('True Positive Rate (Phishing correctly identified)')
plt.title('Receiver Operating Characteristic (ROC) - Random Forest')
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.show()
```

```python
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report

#  Advanced Models Training
advanced_models = {
    "XGBoost": XGBClassifier(n_estimators=200, learning_rate=0.1, max_depth=6, random_state=42),
}

# Training and Evaluating
for name, model in advanced_models.items():
    # Fit model
    model.fit(X_train, y_train)
    
    # Predict
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
    
    print(f"\n" + "="*40)
    print(f"Model: {name}")
    print("="*40)
    
    # Metrics
    print(f"Accuracy: {accuracy_score(y_test, y_pred)*100:.2f}%")
    print(f"F1-Score: {f1_score(y_test, y_pred):.4f}")
    print(f"ROC-AUC:  {roc_auc_score(y_test, y_prob):.4f}")
    
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

# Save the best model for the final plots
best_model = advanced_models["XGBoost"]
```

```python
# Plotting ROC Curve for the Winner
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

fpr, tpr, _ = roc_curve(y_test, best_model.predict_proba(X_test)[:, 1])
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'Best Model AUC = {auc(fpr, tpr):.4f}', color='red', lw=2)
plt.plot([0, 1], [0, 1], '--', color='gray')
plt.title('Final ROC Curve (Phishing Detection)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

#  Prepare the models for comparison
final_comparison = {
    "Random Forest": models["Random Forest"],
    "XGBoost": best_model # This is the XGBClassifier we defined earlier
}

results = []

#  Iterate and calculate metrics for both
for name, model in final_comparison.items():
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
    
    results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "F1-Score": f1_score(y_test, y_pred),
        "ROC-AUC": roc_auc_score(y_test, y_prob)
    })

#  Create a comparison DataFrame
results_df = pd.DataFrame(results).set_index('Model')

print("--- Final Model Comparison Table ---")
print(results_df)

# 4. Visualization: Bar plot for metrics
results_df.plot(kind='bar', figsize=(10, 6), colormap='viridis')
plt.title('Final Comparison: Random Forest vs XGBoost', fontsize=14)
plt.ylabel('Score')
plt.ylim(0.90, 1.0) # Zoom in to see the small differences
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
